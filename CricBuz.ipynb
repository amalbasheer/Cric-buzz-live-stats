```python
# SCORECARD DATA CONNECTED

import psycopg2
def create_scorecard_tables(conn):
    """
    Creates necessary tables for scorecard data in the PostgreSQL database if they don't exist.
    """
    cur = conn.cursor()
    try:
        # Player table first, as others reference it
        cur.execute("""
            CREATE TABLE IF NOT EXISTS players (
                player_id BIGINT PRIMARY KEY,
                name VARCHAR(100) NOT NULL,
                nickname VARCHAR(100),
                is_captain BOOLEAN,
                is_keeper BOOLEAN
            );
        """)

        cur.execute("""
            CREATE TABLE IF NOT EXISTS innings (
                innings_id INT unique,
                score INT,
                wickets INT,
                overs NUMERIC(4,1),
                run_rate NUMERIC(4,2),
                batting_team_name VARCHAR(100),
                batting_team_short_name VARCHAR(10),
                is_declared BOOLEAN,
                is_follow_on BOOLEAN,
                ball_number INT,
                runs_per_ball NUMERIC(4,2)
            );
        """)

        cur.execute("""
            CREATE TABLE IF NOT EXISTS batsmen_stats (
                batsmen_stats_id BIGSERIAL PRIMARY KEY,
                innings_id BIGINT REFERENCES innings(innings_id) ON DELETE CASCADE,
                player_id BIGINT REFERENCES players(player_id),
                balls INT,
                runs INT,
                fours INT,
                sixes INT,
                strike_rate NUMERIC(6,2),
                out_description TEXT
            );
        """)

        cur.execute("""
            CREATE TABLE IF NOT EXISTS bowler_stats (
                bowler_stats_id BIGSERIAL PRIMARY KEY,
                innings_id BIGINT REFERENCES innings(innings_id) ON DELETE CASCADE,
                player_id BIGINT REFERENCES players(player_id),
                overs NUMERIC(4,1),
                maidens INT,
                wickets INT,
                runs_conceded INT,
                economy NUMERIC(4,2),
                dots INT,
                balls INT,
                runs_per_ball NUMERIC(4,2)
            );
        """)

        cur.execute("""
            CREATE TABLE IF NOT EXISTS fall_of_wickets (
                fow_id BIGSERIAL PRIMARY KEY,
                innings_id BIGINT REFERENCES innings(innings_id) ON DELETE CASCADE,
                batsman_id BIGINT, -- No direct foreign key as player_id not always in players table from FOW
                batsman_name VARCHAR(100),
                over_number NUMERIC(4,1),
                runs_at_fall INT,
                ball_number INT
            );
        """)

        cur.execute("""
            CREATE TABLE IF NOT EXISTS extras (
                extras_id BIGSERIAL PRIMARY KEY,
                innings_id BIGINT REFERENCES innings(innings_id) ON DELETE CASCADE,
                leg_byes INT,
                byes INT,
                wides INT,
                no_balls INT,
                penalty INT,
                total INT
            );
        """)

        cur.execute("""
            CREATE TABLE IF NOT EXISTS powerplay (
                powerplay_id BIGSERIAL PRIMARY KEY,
                innings_id BIGINT REFERENCES innings(innings_id) ON DELETE CASCADE,
                powerplay_type VARCHAR(50),
                over_from NUMERIC(4,1),
                over_to NUMERIC(4,1),
                runs_scored INT,
                wickets_lost INT
            );
        """)

        cur.execute("""
            CREATE TABLE IF NOT EXISTS partnerships (
                partnership_id BIGSERIAL PRIMARY KEY,
                innings_id BIGINT REFERENCES innings(innings_id) ON DELETE CASCADE,
                bat1_player_id BIGINT,
                bat1_name VARCHAR(100),
                bat1_runs INT,
                bat1_fours INT,
                bat1_sixes INT,
                bat2_player_id BIGINT,
                bat2_name VARCHAR(100),
                bat2_runs INT,
                bat2_fours INT,
                bat2_sixes INT,
                total_runs INT,
                total_balls INT
            );
        """)
        conn.commit()
        print("All scorecard-related tables created successfully (or already exist).")
    except psycopg2.Error as e:
        conn.rollback()
        print(f"Error creating tables: {e}")
    finally:
        if cur:
            cur.close()

           
```

```python
import psycopg2

def process_and_store_scorecard_data(scorecard_data, conn):
    """
    Processes JSON scorecard_data and upserts it into the DB.
    """
    if conn is None:
        raise ValueError("Invalid database connection.")

    # Turn off autocommit for transactional safety
    conn.autocommit = False
    cur = conn.cursor()

    try:
        for innings_data in scorecard_data.get('scorecard', []):
            # pull fields from JSON
            orig_innings_id = innings_data['inningsid']
            score, wickets = innings_data['score'], innings_data['wickets']
            overs, run_rate = innings_data['overs'], innings_data['runrate']
            bt_name, bt_short = innings_data['batteamname'], innings_data['batteamsname']
            is_decl, is_follow = innings_data['isdeclared'], innings_data['isfollowon']
            ballnbr, rpb = innings_data['ballnbr'], innings_data['rpb']

            # ---- UPSERT innings ----
            cur.execute("""
                INSERT INTO innings
                  (innings_id, score, wickets, overs, run_rate,
                   batting_team_name, batting_team_short_name,
                   is_declared, is_follow_on, ball_number, runs_per_ball)
                VALUES
                  (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
                ON CONFLICT (innings_id) DO UPDATE SET
                  score           = EXCLUDED.score,
                  wickets         = EXCLUDED.wickets,
                  overs           = EXCLUDED.overs,
                  run_rate        = EXCLUDED.run_rate,
                  batting_team_name       = EXCLUDED.batting_team_name,
                  batting_team_short_name = EXCLUDED.batting_team_short_name,
                  is_declared     = EXCLUDED.is_declared,
                  is_follow_on    = EXCLUDED.is_follow_on,
                  ball_number     = EXCLUDED.ball_number,
                  runs_per_ball   = EXCLUDED.runs_per_ball
                RETURNING innings_id;
            """, (
                orig_innings_id, score, wickets, overs, run_rate,
                bt_name, bt_short, is_decl, is_follow, ballnbr, rpb
            ))

            # fetch the real PK back  
            row = cur.fetchone()
            if row:
                innings_id = row[0]
            else:
                # should never happen, but fallback to your JSON ID
                innings_id = orig_innings_id


         # --- Process and Insert 'players' (for batsmen and bowlers) ---
            # Extract players from both batsman and bowler lists to populate the 'players' table
            all_players_in_innings = []
            for batsman in innings_data.get('batsman', []):
                all_players_in_innings.append({
                    'id': batsman.get('id'),
                    'name': batsman.get('name'),
                    'nickname': batsman.get('nickname'),
                    'iscaptain': batsman.get('iscaptain'),
                    'iskeeper': batsman.get('iskeeper')
                })
            for bowler in innings_data.get('bowler', []):
                all_players_in_innings.append({
                    'id': bowler.get('id'),
                    'name': bowler.get('name'),
                    'nickname': bowler.get('nickname'),
                    'iscaptain': bowler.get('iscaptain'),
                    'iskeeper': bowler.get('iskeeper')
                })

            for player_info in all_players_in_innings:
                player_id = player_info.get('id')
                if player_id: # Only insert if player_id is available
                    cur.execute(
                        """
                        INSERT INTO players (player_id, name, nickname, is_captain, is_keeper)
                        VALUES (%s, %s, %s, %s, %s) ON CONFLICT DO NOTHING
                        ;
                        """,
                        (player_id, player_info.get('name'), player_info.get('nickname'),
                         player_info.get('iscaptain'), player_info.get('iskeeper'))
                    )
            print("    -> Processed unique players for Innings")


            # --- 3. Process Batsmen Statistics ---
            for batsman in innings_data.get('batsman', []):
                player_id = batsman.get('id')
                if player_id: # Only insert if player_id is available
                    cur.execute(
                        """
                        INSERT INTO batsmen_stats (innings_id, player_id, balls, runs, fours, sixes, strike_rate, out_description)
                        VALUES (%s, %s, %s, %s, %s, %s, %s, %s) ON CONFLICT DO NOTHING;
                        """,
                        (innings_id, player_id, batsman.get('balls'), batsman.get('runs'),
                         batsman.get('fours'), batsman.get('sixes'),
                         batsman.get('strkrate'), batsman.get('outdec'))
                    )
            print("    -> Processed batsmen stats for Innings")

            # --- 4. Process Bowler Statistics ---
            for bowler in innings_data.get('bowler', []):
                player_id = bowler.get('id')
                if player_id: # Only insert if player_id is available
                    cur.execute(
                        """
                        INSERT INTO bowler_stats (innings_id, player_id, overs, maidens, wickets,
                                                  runs_conceded, economy, dots, balls, runs_per_ball)
                        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s) ON CONFLICT DO NOTHING;
                        """,
                        (innings_id, player_id, bowler.get('overs'), bowler.get('maidens'),
                         bowler.get('wickets'), bowler.get('runs'),  # 'runs' in bowler refers to runs conceded
                         bowler.get('economy'), bowler.get('dots'),
                         bowler.get('balls'), bowler.get('rpb'))
                    )
            print("    -> Processed bowler stats for Innings")

            # --- 5. Process Fall of Wickets ---
            for fow_item in innings_data.get('fow', {}).get('fow', []):
                # Using BIGINT for batsman_id in table, so directly pass here.
                # No ON CONFLICT for these detail tables; assume unique per run.
                cur.execute(
                    """
                    INSERT INTO fall_of_wickets (innings_id, batsman_id, batsman_name, over_number, runs_at_fall, ball_number)
                    VALUES (%s, %s, %s, %s, %s, %s) ON CONFLICT DO NOTHING;
                    """,
                    (innings_id, fow_item.get('batsmanid'), fow_item.get('batsmanname'),
                     fow_item.get('overnbr'), fow_item.get('runs'), fow_item.get('ballnbr'))
                )
            print("    -> Processed fall of wickets for Innings")

            # --- 6. Process Extras ---
            extras_data = innings_data.get('extras', {})
            cur.execute(
                """
                INSERT INTO extras (innings_id, leg_byes, byes, wides, no_balls, penalty, total)
                VALUES (%s, %s, %s, %s, %s, %s, %s) ON CONFLICT DO NOTHING;
                """,
                (innings_id, extras_data.get('legbyes'), extras_data.get('byes'),
                 extras_data.get('wides'), extras_data.get('noballs'),
                 extras_data.get('penalty'), extras_data.get('total'))
            )
            print("    -> Processed extras for Innings")

            # --- 7. Process Powerplay ---
            for pp_item in innings_data.get('pp', {}).get('powerplay', []):
                cur.execute(
                    """
                    INSERT INTO powerplay (innings_id, powerplay_type, over_from, over_to, runs_scored, wickets_lost)
                    VALUES (%s, %s, %s, %s, %s, %s) ON CONFLICT DO NOTHING;
                    """,
                    (innings_id, pp_item.get('pptype'), pp_item.get('ovrfrom'),
                     pp_item.get('ovrto'), pp_item.get('run'), pp_item.get('wickets'))
                )
            print("    -> Processed powerplay for Innings")

            # --- 8. Process Partnerships ---
            for partnership_item in innings_data.get('partnership', {}).get('partnership', []):
                # Ensure bat1id and bat2id are handled
                cur.execute(
                    """
                    INSERT INTO partnerships (innings_id, bat1_player_id, bat1_name, bat1_runs, bat1_fours, bat1_sixes,
                                            bat2_player_id, bat2_name, bat2_runs, bat2_fours, bat2_sixes,
                                            total_runs, total_balls)
                    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s) ON CONFLICT DO NOTHING;
                    """,
                    (innings_id,
                     partnership_item.get('bat1id'), partnership_item.get('bat1name'),
                     partnership_item.get('bat1runs'), partnership_item.get('bat1fours'), partnership_item.get('bat1sixes'),
                     partnership_item.get('bat2id'), partnership_item.get('bat2name'),
                     partnership_item.get('bat2runs'), partnership_item.get('bat2fours'), partnership_item.get('bat2sixes'),
                     partnership_item.get('totalruns'), partnership_item.get('totalballs'))
                )
            print("    -> Processed partnerships for Innings {innings_id}")
        
        conn.commit()
        print(" Scorecard data successfully upserted.")
    except Exception as e:
        conn.rollback()
        print(f" Transaction failed: {e}")
        raise
    finally:
        cur.close()


if __name__ == "__main__":
    # Use your utilis helper to get a connection
    conn = create_connection()
    try:
        # 1. ensure tables exist
        create_scorecard_tables(conn)

        # 2. fetch from API
        scorecard_json = get_scorecard()

        # 3. process & store
        process_and_store_scorecard_data(scorecard_json, conn)
    finally:
        conn.close()
        print("Database connection closed.")
```
